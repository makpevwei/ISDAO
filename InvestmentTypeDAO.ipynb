```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class InvestmentTypeBean:
    itypeId = ""
    iclassId = ""
    itype = ""
    itypeDesc = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class InvestmentTypeDAO:
    def addInvestmentType(self, i=InvestmentTypeBean()):
        try:
     
            sql = "INSERT INTO is_investment_types(i_type_id, i_class_id, i_type, i_type_desc, entry_date) VALUES(%s, %s, %s, %s, %s)"
            data = (i.itypeId, i.iclassId, i.itype, i.itypeDesc, i.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateInvestmentType(self, i=InvestmentTypeBean()):
        try:
            
            sql = "UPDATE is_investment_types SET i_class_id=%s, i_type=%s, i_type_desc=%s, entry_date=%s WHERE i_type_id=%s"
            data = (i.iclassId, i.itype, i.itypeDesc, i.entryDate, i.itypeId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteInvestmentType(self, itypeId):
        try:
            
            sql = "DELETE FROM is_investment_types WHERE i_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (itypeId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getInvestmentType(self, itypeId):
        try:
            
            sql = "SELECT i_type_id id, i_class_id class, i_type type, i_type_desc description, entry_date date FROM is_investment_types WHERE i_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (itypeId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getInvestmentTypeList(self):
        try:
            
            sql = "SELECT i_type_id id, i_class_id class, i_type type, i_type_desc description, entry_date date FROM is_investment_types"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
