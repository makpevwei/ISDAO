```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserBean:
    userId = ""
    user = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserDAO:
    def addUser(self, u=UserBean()):
        try:
     
            sql = "INSERT INTO is_users(user_id, user, entry_date) VALUES(%s, %s, %s)"
            data = (u.userId, u.user, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateUser(self, u=UserBean()):
        try:
            
            sql = "UPDATE is_users SET user=%s, entry_date=%s WHERE user_id=%s"
            data = (u.user, u.entryDate, u.userId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteUser(self, userId):
        try:
            
            sql = "DELETE FROM is_users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (userId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUser(self, userId):
        try:
            
            sql = "SELECT user_id id, user user, entry_date date FROM is_users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserList(self):
        try:
            
            sql = "SELECT user_id id, user user, entry_date date FROM is_users"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
