```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class DisputeLogBean:
    logId = ""
    userId = ""
    cardTypeId = ""
    disputeTypeId = ""
    transactionId = ""
    dispute = ""
    disputeStatus = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class DisputeLogDAO:
    def addDisputeLog(self, d=DisputeLogBean()):
        try:
     
            sql = "INSERT INTO is_dispute_logs(dispute_log_id, user_id, card_type_id, dispute_type_id, transaction_id, dispute, dispute_status, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s)"
            data = (d.logId, d.userId, d.cardTypeId, d.disputeTypeId, d.transactionId, d.dispute, d.disputeStatus, d.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateDisputeLog(self, d=DisputeLogBean()):
        try:
            
            sql = "UPDATE is_dispute_logs SET user_id=%s, card_type_id=%s, dispute_type_id=%s, transaction_id=%s, dispute=%s, dispute_status=%s, entry_date=%s WHERE dispute_log_id=%s"
            data = (d.userId, d.cardTypeId, d.disputeTypeId, d.transactionId, d.dispute, d.disputeStatus, d.entryDate, d.logId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteDisputeLog(self, logId):
        try:
            
            sql = "DELETE FROM is_dispute_logs WHERE dispute_log_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (logId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getDisputeLog(self, logId):
        try:
            
            sql = "SELECT dispute_log_id id, user_id user, card_type_id card, dispute_type_id disputeType, transaction_id transaction, dispute dispute, dispute_status status, entry_date date FROM is_dispute_logs WHERE dispute_log_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (logId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getDisputeLogList(self):
        try:
            
            sql = "SELECT dispute_log_id id, user_id user, card_type_id card, dispute_type_id disputeType, transaction_id transaction, dispute dispute, dispute_status status, entry_date date FROM is_dispute_logs"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
