```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class DisputeTypeBean:
    disputeTypeId = ""
    disputeType = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class DisputeTypeDAO:
    def addDisputeType(self, t=DisputeTypeBean()):
        try:
     
            sql = "INSERT INTO is_dispute_types(dispute_type_id, dispute_type, entry_date) VALUES(%s, %s, %s)"
            data = (t.disputeTypeId, t.disputeType, t.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateDisputeType(self, t=DisputeTypeBean()):
        try:
            
            sql = "UPDATE is_dispute_types SET dispute_type=%s, entry_date=%s WHERE dispute_type_id=%s"
            data = (t.disputeType, t.entryDate, t.disputeTypeId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteDisputeType(self, disputeTypeId):
        try:
            
            sql = "DELETE FROM is_dispute_types WHERE dispute_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (disputeTypeId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getDisputeType(self, disputeTypeId):
        try:
            
            sql = "SELECT dispute_type_id id, dispute_type dispute, entry_date date FROM is_dispute_types WHERE dispute_type_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (disputeTypeId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getDisputeTypeList(self):
        try:
            
            sql = "SELECT dispute_type_id id, dispute_type dispute, entry_date date FROM is_dispute_types"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
