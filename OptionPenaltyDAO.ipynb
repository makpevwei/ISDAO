```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class OptionPenaltyBean:
    penaltyId = ""
    penalty = ""
    penaltyDesc = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class OptionPenaltyDAO:
    def addOptionPenalty(self, p=OptionPenaltyBean()):
        try:
     
            sql = "INSERT INTO is_option_penalties(penalty_id, penalty, penalty_desc, entry_date) VALUES(%s, %s, %s, %s)"
            data = (p.penaltyId, p.penalty, p.penaltyDesc,  p.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateOptionPenalty(self, p=OptionPenaltyBean()):
        try:
            
            sql = "UPDATE is_option_penalties SET penalty=%s, penalty_desc=%s, entry_date=%s WHERE penalty_id=%s"
            data = (p.penalty, p.penaltyDesc, p.entryDate, p.penaltyId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteOptionPenalty(self, penaltyId):
        try:
            
            sql = "DELETE FROM is_option_penalties WHERE penalty_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (penaltyId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getOptionPenalty(self, penaltyId):
        try:
            
            sql = "SELECT penalty_id id, penalty penalty, penalty_desc, entry_date date FROM is_option_penalties WHERE penalty_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (penaltyId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getOptionPenaltyList(self):
        try:
            
            sql = "SELECT penalty_id id, penalty penalty, penalty_desc, entry_date date FROM is_option_penalties"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
