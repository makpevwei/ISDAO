```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class InsuranceCompanyBean:
    insuranceId = ""
    insuranceCompany = ""
    contactPerson = ""
    contactMobile = ""
    contactEmail = ""
    rcNumber = ""
    location = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class InsuranceCompanyDAO:
    def addInsuranceCompany(self, i=InsuranceCompanyBean()):
        try:
     
            sql = "INSERT INTO is_insurance_companies(insurance_id, insurance_company, contact_person, contact_mobile, contact_email, rc_number, head_office_location, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s)"
            data = (i.insuranceId, i.insuranceCompany, i.contactPerson, i.contactMobile, i.contactEmail, i.rcNumber, i.location, i.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateInsuranceCompany(self, i=InsuranceCompanyBean()):
        try:
            
            sql = "UPDATE is_insurance_companies SET insurance_company=%s, contact_person=%s, contact_mobile=%s, contact_email=%s, rc_number=%s, head_office_location=%s, entry_date=%s WHERE insurance_id=%s"
            data = (i.insuranceCompany, i.contactPerson, i.contactMobile, i.contactEmail, i.rcNumber, i.location, i.entryDate, i.insuranceId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            ()
            

            
    def deleteInsuranceCompany(self, insuranceId):
        try:
            
            sql = "DELETE FROM is_insurance_companies WHERE insurance_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (insuranceId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getInsuranceCompany(self, insuranceId):
        try:
            
            sql = "SELECT insurance_id id, insurance_company company,  contact_person person, contact_mobile mobile, contact_email email, rc_number rc, head_office_location location, entry_date date FROM is_insurance_companies WHERE insurance_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (insuranceId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getInsuranceCompanyList(self):
        try:
            
            sql = "SELECT insurance_id id, insurance_company company,  contact_person person, contact_mobile mobile, contact_email email, rc_number rc, head_office_location location, entry_date date FROM is_insurance_companies"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
