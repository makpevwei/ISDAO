```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class TransactionBean:
    transactionId = ""
    transaction = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class TransactionDAO:
    def addTransaction(self, t=TransactionBean()):
        try:
     
            sql = "INSERT INTO is_transactions(transaction_id, transaction, entry_date) VALUES(%s, %s, %s)"
            data = (t.transactionId, t.transaction, t.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateTransaction(self, t=TransactionBean()):
        try:
            
            sql = "UPDATE transactions SET is_transaction=%s, entry_date=%s WHERE transaction_id=%s"
            data = (t.transaction, t.entryDate, t.transactionId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteTransaction(self, transactionId):
        try:
            
            sql = "DELETE FROM is_transactions WHERE transaction_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (transactionId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getTransaction(self, transactionId):
        try:
            
            sql = "SELECT transaction_id id, transaction transaction, entry_date date FROM is_transactions WHERE transaction_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (transactionId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getTransactionList(self):
        try:
            
            sql = "SELECT transaction_id id, transaction transaction, entry_date date FROM is_transactions"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
