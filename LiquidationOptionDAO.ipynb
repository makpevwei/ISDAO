```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class LiquidationOptionBean:
    optionId = ""
    liquidOption = ""
    optionDesc = ""
    penaltyId = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class LiquidationOptionDAO:
    def addLiquidationOption(self, l=LiquidationOptionBean()):
        try:
     
            sql = "INSERT INTO is_liquidation_options(option_id, liquidation_option, option_desc, penalty_id, entry_date) VALUES(%s, %s, %s, %s, %s)"
            data = (l.optionId, l.liquidOption, l.optionDesc, l.penaltyId,  l.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def updateLiquidationOption(self, l=LiquidationOptionBean()):
        try:
            
            sql = "UPDATE is_liquidation_options SET liquidation_option=%s, option_desc=%s, penalty_id=%s, entry_date=%s WHERE option_id=%s"
            data = (l.liquidOption, l.optionDesc, l.penaltyId,  l.entryDate, l.optionId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, data)
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            

            
    def deleteLiquidationOption(self, optionId):
        try:
            
            sql = "DELETE FROM is_liquidation_options WHERE option_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            rowCount = cursor.execute(sql, (optionId,))
            conn.commit()
            
            if rowCount > 0:
                return 0
            return 1
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getLiquidationOption(self, optionId):
        try:
            
            sql = "SELECT option_id id, liquidation_option liquidation, option_desc description, penalty_id penalty, entry_date date FROM is_liquidation_options WHERE option_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (optionId,))
            record = cursor.fetchone()
            
            return record
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getLiquidationOptionList(self):
        try:
            
            sql = "SELECT option_id id, liquidation_option liquidation, option_desc description, penalty_id penalty, entry_date date FROM is_liquidation_options"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            return records
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
